```python
from stanfordcorenlp import StanfordCoreNLP
import pandas as pd
import json
import os
import re
```

```python
nlp = StanfordCoreNLP(r'C:\Users\user\stanfordcorenlp', lang='zh')
```

```python
data = pd.read_csv('515K hotel dataset.csv')
```

```python
# remove = '[0-9"#$%&\'()*+-/<=>?@?[\\]^_`{|}~]+'  #
training_list=[[] for i in range(13)]
def read_training():
    i=0
    training_id = open('./training.txt','r')
    for line in training_id:
        training_list[i] = line.strip('\n').strip(' ').split(' ')
        i +=1
    training_id.close()
read_training()

def read_train_corpus(corpus_id): 
    training_file = open('./IRTM/'+str(corpus_id)+'.txt','r')
    train = training_file.read()
    training_file.close()
    return train


for c in range(13):
    for file in training_list[0][1:]:
        print('-------------------------------review-----------------------------------------')
        corpus = read_train_corpus(file)
        negativeResults = 0
        neutralResults = 0
        positiveResults = 0
        for k in corpus.split("."):
            sentence = re.sub(r"[^a-zA-Z0-9]+", ' ', k)
            result = nlp.annotate(sentence,properties={'annotators': 'sentiment','outputFormat': 'json','timeout': 1000,})
            jsonSentiment = json.loads(result)
            for s in jsonSentiment["sentences"]:
                print ( " ".join([word["word"] for word in s["tokens"]]) + " => " + str(s["sentimentValue"]) + " = "+ s["sentiment"])
                if(s["sentiment"] == "Negative"):
                    negativeResults = negativeResults + 1
                elif(s["sentiment"] == "Neutral"):
                    neutralResults = neutralResults + 1
                elif(s["sentiment"] == "Positive"):
                    positiveResults = positiveResults + 1
        print(negativeResults)
        print(positiveResults)
        print(neutralResults)
        if negativeResults > positiveResults and negativeResults > neutralResults:
            print('negative')
        elif positiveResults > negativeResults and positiveResults > neutralResults:
            print('positive')
        elif neutralResults > negativeResults and neutralResults > positiveResults:
            print('neutral')
```

```python
data.dropna(inplace = True)
data.drop('Unnamed: 0', 1, inplace =True)
data.reset_index(drop = True, inplace = True)
```

```python
#One-side Reviews: Remove the review with "No Negative"/"No Positive". 
data['NegativeReview'].replace('No Negative', "", inplace = True)
data['PositiveReview'].replace('No Positive', "", inplace = True)
```

```python
corpus = data.NegativeReview + data.PositiveReview
data.insert(0, "Review", corpus.str.lower())
```
